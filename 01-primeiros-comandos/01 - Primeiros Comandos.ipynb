## Tipos de Dados

str - string (texto)
A principal característica é a utilização de aspas ''  ""

int - número inteiro
A principal característica é um número sem aspas e sem casas decimais

float - número decimal
A principal característica é um número com casas decimais, usando ponto '.' como separador.

bool - booleano (verdadeiro/falso) (True/False)

Comando type()

```python
# verificação do tipo de dados

nome = 'Thiago'
idade = 36
teste = False
pi = 3.14159

print(type(nome))
print(type(idade))
print(type(teste))
print(type(pi))
```

## Exercícios

```python
# 1) Gerar um script Python que leia o Nome da Pessoa e mostre na tela, juntamente com uma mensagem de Boas Vindas.
# Resultado: Olá fulano. Prazer em te conhecer!

nome = input('Qual é o seu nome? ')
print('Olá',nome,'. Prazer em te conhecer!')
print('Olá '+nome+'. Prazer em te conhecer!')
```

```python
# 2) Crie um script Python que leia o Dia, Mês e o Ano do nascimento de uma pessoa e mostre na tela a informação formatada conforme abaixo:
# Resultado: Você nasceu em 11 de abril de 1970.

dia = input('Digite o dia de nascimento: ')
mes = input('Digite o mês de nascimento, por extenso: ')
ano = input('Digite o ano de nascimento, com 4 dígitos: ')

print('Você nasceu em',dia,'de',mes,'de',ano)
print('Você nasceu em '+dia+' de '+mes+' de '+ano)
```

```python
# 3) Crie um script Python que leia dois números e mostre na tela a soma entre eles
# Resultado: 4 + 4 = 8

# int() - transforma um input em número inteiro, caso seja um número inteiro
# float() - transformar um input em um número decimal, caso seja um número

num1 = int(input('Digite o primeiro número: '))
num2 = int(input('Digite o segundo número: '))

print(type(num1))
print(type(num2))

print('O resultado da soma é:',num1+num2)
print(f'{num1} + {num2} = {num1+num2}')
```

## Formatação de Saída

```python
# f-string - Permite algumas formatações no texto de exibição para o usuário

dia = input('Digite o dia de nascimento: ')
mes = input('Digite o mês de nascimento, por extenso: ')
ano = input('Digite o ano de nascimento, com 4 dígitos: ')

print('Você nasceu em',dia,'de',mes,'de',ano)
print('Você nasceu em '+dia+' de '+mes+' de '+ano)

print(f'Você nasceu em {dia} de {mes} de {ano}.')
```

```python
# Exercício:
# Monte um script python que receba o peso e altura de uma pessoa
# e mostre na tela o seu IMC
# imc = peso dividido pela altura ao quadrado
# imc = peso / (altura * altura)

peso = float(input('Informe o seu peso: '))
altura = float(input('Informe a sua altura: '))

imc = peso / (altura * altura)
print(f'O seu IMC é {imc:.2f}.')
```
