## Precedência

```python
# exemplos de precedências dos operadores

print(2+3*2)         # primeiro multiplicação, depois soma
print((2+3)*2)       # primeiro soma, depois multiplica
print(2*5/2**2)      # primeiro exponenciação, depois multiplica, depois divide
print(2*(5/2**2))    # primeiro exponenciação, depois divisão, depois multiplica
print((2*(8/(2-1)))) # primeiro subtração, depois divisão, depois multiplica
print(8/2*2)         # primeiro divisão, depois multiplica
print(2**2**3)       # primeiro a exponenciação mais a direita, depois a esquerda
```

```python
# Exercício 1:
# Conversor de Temperatura: temperatura em Celsius. Ex: 25ºc
# Converta para Fahrenheit usando a fórmula: (0°C * 9/5) + 32 = 32°F
# Imprima a temperatura em Fahrenheit.

temperatura = float(input('Digite a temperatura em ºC: '))
fahrenheit = temperatura*(9/5)+32

print(f'A temperatura de {temperatura}ºC em Fahrenheit é de {fahrenheit:.2f}ºF.')
```

```python
# Exercício 2:
# Desconto: Informe o valor de um produto e o percentual de desconto.
# Calcule o valor do desconto e o preço final. Exiba os resultados.

valor = float(input('Digite o valor do produto: '))
desconto = int(input('Digite a % de desconto: '))
vrdesconto = valor * (desconto/100)
vrfinal = valor-vrdesconto

print(f'Valor do Produto: R$ {valor:,.2f}')
print(f'% de Desconto: {desconto}%')
print(f'Valor do Desconto: R$ {vrdesconto:,.2f}')
print(f'Valor Final: R$ {vrfinal:,.2f}')
```

```python
# Exercício 3:
# Velocidade Média: Informe a distância percorrida e o tempo gasto.
# Calcule a velocidade média usando a fórmula: VM = Distância / Tempo

distancia = float(input('Digite a distância em km: '))
tempo = float(input('Digite o tempo em horas: '))
vm = distancia/tempo

print(f'Velocidade Média: {vm:.2f} km/h.')
```
