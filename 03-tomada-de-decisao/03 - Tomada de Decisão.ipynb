```python
# Exercício : Aprovação de Empréstimo
# Objetivo: Verificar aprovação de empréstimo com novos critérios:
# Valor do empréstimo (R$)
# Prazo (meses)
# Cálculo Automático: Valor da parcela = (valor do empréstimo / prazo)
# Requisitos para aprovação:
# Renda mínima: R$ 3.000
# Score de crédito: >= 500
# Parcela máxima: <= 35% da renda
# Número de parcelas: Máx 36

renda = float(input('Digite o valor da sua renda: '))
score = float(input('Digite o score informado pelo SERASA: '))
if renda >= 3000 and score >= 500:
    renda35 = renda * (35/100)
    print(f'Você pode comprometer até R$ {renda35:.2f} da sua renda mensal! ')

    vr_emprestimo = float(input('Informe o valor do empréstimo: '))
    qt_parcelas = int(input('Informe a quantidade de parcelas: (Máx:36) '))

    if qt_parcelas <= 36:
        vr_parcela = vr_emprestimo / qt_parcelas

        if vr_parcela <= renda35:
            print(f'Empréstimo aprovado em {qt_parcelas}x de R$ {vr_parcela:.2f}')
        else:
            print(f'Valor da parcela maior do que o valor permitido! ')
            print(f'O valor da parcela escolhida foi R$ {vr_parcela:.2f}')
    else:
        print('Número de parcelas superior ao permitido!')
else:
    print('Empréstimo negado!')
```

```python
# Exercício: Cálculo de frete

# Você precisa programar a lógica de cálculo de frete para uma loja online.
# O custo do frete não é fixo; ele depende do valor da compra, do peso do
# produto e da região de entrega.

# Regras:
# Frete Grátis: Compras acima de R$ 200,00 têm frete grátis
# Custo por Região (para compras até R$ 200,00):
#        Capital:  Custo R$ 10,00.
#        Interior: Custo base de R$ 15,00.
#
# Adicional por Peso (para compras até R$ 200,00):
#        Se o pacote pesar mais de 2 kg, um adicional de R$ 5,00 é somado
#
# Frete Expresso:
#        O cliente pode optar por um frete expresso, que custa R$ 8,00 a mais.
#        Esse valor é adicionado ao custo total do frete
#       (mesmo que seja frete grátis, o expresso é cobrado)

valor_compra = float(input('Digite o valor da compra: \n'))
frete = 0

if valor_compra < 200:
    peso = float(input('Digite o peso: \n'))
    if peso > 2:
        frete = frete + 5

    regiao = input('1 - Capital \n2 - Interior \n').lower()
    if regiao == '1' or regiao == 'capital':
        frete = frete + 10 # frete += 10
    else:
        frete = frete + 15 # frete += 15

expresso = input('Deseja frete expresso? \n1 - Sim\n2 - Não \n').lower()
if expresso == '1' or expresso == 'sim':
    frete = frete + 8

valor_total = valor_compra + frete

print(f'Valor da Compra = R$ {valor_compra:.2f}')
print(f'Valor do Frete = R$ {frete:.2f}')
print(f'Valor da Total = R$ {valor_total:.2f}')
```

```python
# Faça um programa que leia o Nome do aluno, Nome da Matéria,
# as Notas separadas do 1º Bimestre, 2º Bimestre, 3º Bimestre e 4º Bimestre.
# Em seguida calcule a Média.

# Apresente na Tela as seguintes informações:

# ----------------------
#     Boletim Escolar
# ----------------------
# Matéria:
# Nome do aluno:
# Média:
# Situação:
# ----------------------

# Matéria: (Mostre o nome da matéria)
# Nome do aluno : (Mostre o nome do aluno)
# Média Calculada: (Mostre a média e se quiser as notas)
# Situação: (Mostre Aprovado, Recuperação ou Reprovado)
# Média >= 8 --> aprovado   >= 5 --> recuperação  < 5 -- > reprovado

nome = input('Digite o nome do aluno: ')
materia = input('Informe a matéria: ')

nota1 = float(input('Digite a 1ª nota: '))
nota2 = float(input('Digite a 2ª nota: '))
nota3 = float(input('Digite a 3ª nota: '))
nota4 = float(input('Digite a 4ª nota: '))

media = (nota1 + nota2 + nota3 + nota4) / 4

if media >= 8:
    situacao = 'Aprovado 😁'
elif media < 5:
    situacao = 'Reprovado 😞'
else:
    situacao = 'Recuperação 😐'

print(f'\nBoletim Escolar: \n')
print(f'Nome do Aluno: {nome}')
print(f'Matéria: {materia}')
print(f'Média: {media}')
print(f'Situação: {situacao}')
```

```python
# Crie um programa que faça uma competição de Jokenpô entre 2 jogadores:
# PEDRA ganha de TESOURA
# TESOURA ganha de PAPEL
# PAPEL ganha de PEDRA

from google.colab import output

player1 = input('Jogador 1: Pedra ✊, Papel ✋, Tesoura ✌️:').capitalize()
output.clear()
player2 = input('Jogador 2: Pedra ✊, Papel ✋, Tesoura ✌️:').capitalize()
output.clear()

if player1 == player2:
    print('Empate')
elif (player1 == 'Pedra' and player2 == 'Tesoura') or (player1 == 'Tesoura' and player2 == 'Papel') or (player1 == 'Papel' and player2 == 'Pedra'):
    print('Jogador 1 venceu!')
elif (player2 == 'Pedra' and player1 == 'Tesoura') or (player2 == 'Tesoura' and player1 == 'Papel') or (player2 == 'Papel' and player1 == 'Pedra'):
    print('Jogador 2 venceu!')
else:
    print('Jogada inválida!')
print(f'Jogador1 --> {player1}  x {player2} <-- Jogador2')
```

```python
# Crie um programa que faça uma competição de Jokenpô entre você e a IA:
# PEDRA ganha de TESOURA
# TESOURA ganha de PAPEL
# PAPEL ganha de PEDRA

from google.colab import output
from random import choice

opcao = ['Pedra','Papel','Tesoura']
atom = choice(opcao)

# print(atom)

humano = input('Escolha: Pedra ✊, Papel ✋, Tesoura ✌️: ').capitalize()

if humano == atom:
    print('Empate')
elif (humano == 'Pedra' and atom == 'Tesoura') or (humano == 'Tesoura' and atom == 'Papel') or (humano == 'Papel' and atom == 'Pedra'):
    print('Humano venceu!')
elif (atom == 'Pedra' and humano == 'Tesoura') or (atom == 'Tesoura' and humano == 'Papel') or (atom == 'Papel' and humano == 'Pedra'):
    print('Atom venceu!')
else:
    print('Jogada inválida!')
print(f'Humano --> {humano}  x {atom} <-- Atom')
```
