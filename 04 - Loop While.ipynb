```python
# Faça um programa que peça dois números
# e uma operação matemática (+, -, *, /)
# mostrar o resultado (executar 4x (1x para cada operação))

from google.colab import output
from time import sleep

i = 1

while i <= 4:
    num1 = float(input('Digite o primeiro número: '))
    num2 = float(input('Digite o segundo número: '))

    op = input("""
        [1] Soma
        [2] Subtração
        [3] Multiplicação
        [4] Divisão
    """)

    if op == '1':
        print(num1+num2)
    elif op == '2':
        print(num1-num2)
    elif op == '3':
        print(num1*num2)
    else:
        if num2 == 0:
            print('Não é possível dividir por 0!')
        else:
            print(num1/num2)
    i = i+1
    sleep(6)
    output.clear()

print('Fim do programa após 4 execuções!')

```

```python
# Faça ATOM e ZEUS jogarem 20x, armazene o placar:
# PEDRA ganha de TESOURA
# TESOURA ganha de PAPEL
# PAPEL ganha de PEDRA

from google.colab import output
from random import choice
from time import sleep

vitoria_atom = 0
vitoria_zeus = 0
empates = 0
i = 1
opcao = ['Pedra','Papel','Tesoura']

while i <= 1000000:
    atom = choice(opcao)
    zeus = choice(opcao)

    if zeus == atom:
        # print('Empate')
        empates += 1
    elif (zeus == 'Pedra' and atom == 'Tesoura') or (zeus == 'Tesoura' and atom == 'Papel') or (zeus == 'Papel' and atom == 'Pedra'):
        # print('Zeus venceu!')
        vitoria_zeus += 1
    elif (atom == 'Pedra' and zeus == 'Tesoura') or (atom == 'Tesoura' and zeus == 'Papel') or (atom == 'Papel' and zeus == 'Pedra'):
        # print('Atom venceu!')
        vitoria_atom += 1
    else:
        print('Jogada inválida!')

    i += 1
    # print(f'Zeus --> {zeus}  x {atom} <-- Atom')
    # sleep(2)

print('Placares: \n')
print(f'Vitórias Zeus: {vitoria_zeus}')
print(f'Vitórias Atom: {vitoria_atom}')
print(f'Empates: {empates}')
```

```python
# Exercício:
# Desenvolver um programa que vai solicitar diversos números
# e após o comando de saída ele vai exibir qual o maior
# número e qual o menor número digitado, bem como a quantidade
# de números digitados.

quantidade = 0
maior = float('-inf')
menor = float('inf')

while True:
    numero = input('Digite um número qualquer ("S" para sair): ').lower()

    if numero == 's':
        break
    else:
        numero = float(numero)
        quantidade += 1

        if numero >= maior:
            maior = numero
        if numero <= menor:
            menor = numero
print(f'Números digitados: {quantidade}')
print(f'O maior número foi: {maior}')
print(f'O menor número foi: {menor}')
```

```python
# Faça um programa que escolha uma letra qualquer do alfabeto
# tente acertar a letra

from random import choice
from time import sleep
from google.colab import output
import string

alfabeto_minusculo = list(string.ascii_lowercase)

letra_secreta = choice(alfabeto_minusculo)
# print(letra_secreta)
while True:
    letra = input('Digite uma letra de "a" até "z": ').lower()

    if letra == letra_secreta:
        print('Você acertou a letra!')
        break
    elif letra < letra_secreta:
        print('Esta letra é menor!')
    elif letra > letra_secreta:
        print('Esta letra é maior!')
    else:
        print('Entrada inválida')
    sleep(4)
    output.clear()

print('Fim do programa!')
```
