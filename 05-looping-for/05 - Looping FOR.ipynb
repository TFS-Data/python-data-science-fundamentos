```python
# #Desenvolva um programa que leia dez números inteiros e mostre a soma apenas
# daqueles que forem pares. Se o valor digitado for ímpar desconsidere-o.
# contar os números pares digitados.

par = 0
contador = 0

for i in range(6): # 0 a 5
    num = int(input(f'Digite o {i+1}º número: '))
    if num % 2 == 0:
        par += num
        contador += 1

print(f'Soma dos números pares: {par}')
print(f'Quantidade de números pares: {contador}')
```

```python
# Faça um programa que leia o peso de cinco pessoas.
# No final, mostre qual foi o maior e o menor peso lidos.

maior = float('-inf')
menor = float('inf')

for i in range(1, 2):
    peso = float(input(f'Digite o {i}º peso: '))

    if peso > maior:
        maior = peso
    if peso < menor:
        menor = peso

print(f'Maior peso: {maior:.1f} Kg')
print(f'Menor peso: {menor:.1f} Kg')
```

```python
# Faça um programa que peça a idade de 15 pessoas
# mostre quantas são maiores de idade e quantas são menores de idade
from time import sleep

maiores = 0
menores = 0
ano_atual = 2026

for i in range(1,7): # 0 a 5
    ano = int(input(f'Digite o ano de nascimento da pessoa {i}: '))
    idade = ano_atual - ano

    if idade >= 18:
        maiores += 1
    else:
        menores += 1

    print(f'Idade: {idade} ')
    sleep(1)

print(f'Quantidade de maiores: {maiores}')
print(f'Quantidade de menores: {menores}')
```
