```python
# Manipulação de Dados

import pandas as pd
```

```python
# DataFrames - df
# df_exemplo ou df_vendas

# importando um arquivo csv separado por vírgula

df = pd.read_csv('/content/drive/MyDrive/!Data Python 2603/base_vendas.csv')
print(df)
```

```python
# Display - Mostra as 5 primeiras linhas e as 5 últimas linhas
# caso o arquivo for pequeno, ele exibe todas

display(df)
```

```python
# Na primeira vez que abrimos, ele entende que está separado por vírgula
# Para escolhermos o separador, podemos usar:
# delimiter='separador'   ou  sep='separador'

# importando um arquivo csv separado por ponto e vírgula

df = pd.read_csv('/content/drive/MyDrive/!Data Python 2603/base_vendas_ponto_virgula.csv', sep=';')
display(df)
```

```python
# importando um arquivo csv separado por tabulação

df = pd.read_csv('/content/drive/MyDrive/!Data Python 2603/base_vendas_tab.tsv', sep='\t')
display(df)
```

```python
# importando um arquivo h5

df = pd.read_hdf('/content/drive/MyDrive/!Data Python 2603/base_vendas.h5')
display(df)
```

```python
# importando um arquivo json

df = pd.read_json('/content/drive/MyDrive/!Data Python 2603/base_vendas.json')
display(df)
```

```python
# importando um arquivo xml

df = pd.read_xml('/content/drive/MyDrive/!Data Python 2603/base_vendas.xml')
display(df)
```

```python
# importando um arquivo excel
# por padrão ele lê a primeira pasta

df = pd.read_excel('/content/drive/MyDrive/!Data Python 2603/Vendas.xlsx')
display(df)
```

```python
# para ler outra pasta, usamos o parâmetro 'sheet_name=nome da pasta'

df = pd.read_excel('/content/drive/MyDrive/!Data Python 2603/Vendas.xlsx', sheet_name='base_vendas')
display(df)
```
