O CRISP-DM (Cross-Industry Standard Process for Data Mining) é a metodologia mais utilizada para organizar projetos de ciência de dados e mineração de dados, focada em transformar dados em valor de negócio através de seis etapas cíclicas: compreensão do negócio, entendimento dos dados, preparação, modelagem, avaliação e implantação.
As 6 Etapas do CRISP-DM:

Compreensão do Negócio (Business Understanding): Definir objetivos, requisitos e critérios de sucesso do projeto sob a perspectiva de negócio.

Entendimento dos Dados (Data Understanding): Coletar, explorar e identificar a qualidade dos dados disponíveis.

Preparação dos Dados (Data Preparation): Limpar, transformar e selecionar os dados para modelagem (fase geralmente mais demorada).

Modelagem (Modeling): Aplicar técnicas de machine learning ou estatística para criar o modelo.

Avaliação (Evaluation): Verificar se o modelo atende aos objetivos de negócio definidos inicialmente.

Implantação (Deployment): Colocar o modelo em produção, monitorar e documentar os resultados.
