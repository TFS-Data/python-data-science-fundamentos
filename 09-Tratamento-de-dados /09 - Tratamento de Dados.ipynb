```python
# alterando o tipo de dado de string (object) para datetime (ao trazer o arquivo)

df_vendas = pd.read_csv('/content/drive/MyDrive/!Data Python 2603/base_vendas.csv', parse_dates=['dt_venda'], dayfirst=True)
```

```python
# alterando o tipo de dado de string (object) para datetime (maior controle)

df_vendas['dt_entrega'] = pd.to_datetime(df_vendas['dt_entrega'], format='%d/%m/%Y %H:%M')
```

```python
# trocando a vírgula por ponto dentro da string (object)

df_vendas['preco_custo'] = df_vendas['preco_custo'].str.replace(',', '.')
df_vendas['preco_custo'].head(8)
```

```python
# alterando o tipo de dado de string (object) para decimal (float)

df_vendas['preco_custo'] = df_vendas['preco_custo'].astype(float)
df_vendas['preco_custo'].head(8)
```

```python
df_vendas.info()
```

```python
#  trocando a vírgula por ponto dentro da string (object) e
# alterando o tipo de dado para decimal (float) ao mesmo tempo

df_vendas['valor_unitario'] = df_vendas['valor_unitario'].str.replace(',', '.').astype(float)
df_vendas['valor_unitario'].head(8)
```

```python
df_vendas.info()
```

```python
# eliminando colunas do DataFrame

df_vendas.drop(columns=['matricula_funcionario','codigo_produto'], inplace=True)
```

```python
df_vendas
```

```python
# ENGENHARIA DE DADOS / DATA ENGINEER
# Criando novas colunas no DataFrame
# 1 ) coluna --> valor_final

df_vendas['valor_final'] = df_vendas['quantidade'] * df_vendas['valor_unitario']
df_vendas
```

```python
# 2 ) coluna --> lucro_sem_comissao (lucro sem deduzir a comissão do vendedor)

# df_vendas.insert(9, 'lucro_sem_comissao', (df_vendas['valor_unitario'] - df_vendas['preco_custo']) * df_vendas['quantidade'])

df_vendas['lucro_sem_comissao'] = (df_vendas['valor_unitario'] - df_vendas['preco_custo']) * df_vendas['quantidade']
df_vendas
```

```python
# 3 ) coluna --> valor_comissao

df_vendas['valor_comissao'] = (df_vendas['valor_final'] * df_vendas['comissao']).round(2)
df_vendas
```

```python
# 4 ) coluna --> lucro_real

df_vendas['lucro_real'] = df_vendas['lucro_sem_comissao'] - df_vendas['valor_comissao']
df_vendas
```

```python
df_vendas
```

```python
# 5 ) colunas --> margem_lucro

df_vendas['margem_lucro'] = ((df_vendas['valor_unitario'] - df_vendas['preco_custo'])/ df_vendas['valor_unitario']).round(2)
df_vendas
```

```python
df_vendas['margem_lucro2'] = df_vendas['margem_lucro'] * 100
```
